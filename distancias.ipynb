## Distancias en topografía

En este notebook se calcula la distancia geométrica, la distancia reducida y el desnivel a partir de las coordenadas de dos puntos.

**Introduce las coordenadas de la estación:**

```python
X1=100
Y1=100
Z1=100
```

**Introduce las coordenadas del punto observado:**

```python
X2=110
Y2=95
Z2=120
```

### Representación en el espacio 3D

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt


fig = plt.figure(figsize=plt.figaspect(2.))
fig.suptitle('Vista en 3D')
ax = fig.add_subplot(211, projection='3d')

x =[X1,X2]
y =[Y1,Y2]
z =[Z1,Z2]

ax.plot(x, y, z, c='r', marker='o')

plt.show();
```

### Vista en planta

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
fig = plt.figure(figsize=plt.figaspect(1.))
fig.suptitle('Vista en planta')

ax.set_xlabel('X');
ax.set_ylabel('Y');


ax = fig.add_subplot(111);

ax.plot(x, y, c='r', marker='o');

ax.set_xlabel('X');
ax.set_ylabel('Y');
plt.show()
```

```python
import math
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

incX=X2-X1
incY=Y2-Y1
incZ=Z2-Z1

dist=math.sqrt(incX**2+incY**2)

fig = plt.figure(figsize=plt.figaspect(1.))
fig.suptitle('Distancias en topografía')
ax = fig.add_subplot(111);
x=[0,dist]
y=[0,incZ]
ax.plot(x, y, c='r', marker='o',label='Distancia geométrica');
x=[0,dist]
y=[0,0]
ax.plot(x, y, c='b', marker='o',label='Distancia reducida');
x=[dist,dist]
y=[0,incZ]
ax.plot(x, y, c='g', marker='o',label='Desnivel');
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
ax.set_xlabel('Distancia reducida');
ax.set_ylabel('Desnivel');

plt.show()
```

```python
incX=X2-X1;
incY=Y2-Y1;
incZ=Z2-Z1;

dr=math.sqrt(incX**2+incY**2);
dg=math.sqrt(incX**2+incY**2+incZ**2);
dz=incZ;
```

### Distancias

```python
print ('Distancia geométrica', '%.3f'%(dg),'m')
print ('Distancia reducida', '%.3f'%(dr),'m')
print ('Desnivel', '%.3f'%(dz),'m')
```
